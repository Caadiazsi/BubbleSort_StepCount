```python
import numpy as np
import itertools
import matplotlib.pyplot as plt
from collections import Counter
```

```python
print("Ingrese el tamaño del arreglo:")
A = int(input())
```

```python
array = itertools.permutations(range(A))
l_pasos, l_comparaciones, l_intercambios = [], [], []
for x in array:
    a = np.array(x)
    pasos,comparaciones,intercambios = 0,0,0
    for y in range(len(a)-1,0,-1):
        pasos +=1
        for i in range(y):
            pasos +=1
            comparaciones += 1
            if a[i]>a[i+1]:
                a[i], a[i+1] = a[i+1], a[i]
                pasos +=1
                intercambios +=1
        l_pasos.append(pasos)
        l_comparaciones.append(comparaciones)
        l_intercambios.append(intercambios)
```

```python
c_pasos = Counter(l_pasos)
c_comparaciones = Counter(l_comparaciones)
c_intercambios = Counter(l_intercambios)
Tamano_Plot = plt.rcParams["figure.figsize"]
Tamano_Plot[0] = 20
Tamano_Plot[1] = 10
plt.rcParams["figure.figsize"] = Tamano_Plot
```

```python
print ('Numero promedio de pasos para ordenar un arreglo de tamaño {}:'.format(A), np.mean(l_pasos))
print ('Numero minimo de pasos para ordenar un arreglo de tamaño {}:'.format(A), min(l_pasos))
print ('Numero maximo de pasos para ordenar un arreglo de tamaño {}:'.format(A), max(l_pasos))
    
plt.bar(range(len(c_pasos)), list(c_pasos.values()))
plt.xticks(range(len(c_pasos)), list(c_pasos.keys()))
plt.xlabel('Numero de pasos')
plt.ylabel('Cantidad')
plt.title('Pasos para n = {}'.format(A))    
print('\nGrafica:')
plt.show()    
print ('Numero promedio de comparaciones para ordenar un arreglo de tamaño {}:'.format(A),  np.mean(l_comparaciones))
print ('Numero minimo de comparaciones para ordenar un arreglo de tamaño {}:'.format(A), min(l_comparaciones))
print ('Numero maximo de comparaciones para ordenar un arreglo de tamaño {}:'.format(A), max(l_comparaciones) )   
plt.bar(range(len(c_comparaciones)), list(c_comparaciones.values()))
plt.xticks(range(len(c_comparaciones)), list(c_comparaciones.keys()))
plt.xlabel('Numero de comparaciones')
plt.ylabel('Cantidad')
plt.title('Comparaciones para n = {}'.format(A))    
print('\nGrafica:')
plt.show()        
print ('Numero promedio de intercambios para ordenar un arreglo de tamaño {}:'.format(A),  np.mean(l_intercambios))
print ('Numero minimo de intercambios para ordenar un arreglo de tamaño {}:'.format(A), min(l_intercambios))   
print ('Numero maximo de intercambios para ordenar un arreglo de tamaño {}:'.format(A), max(l_intercambios)) 
plt.bar(range(len(c_intercambios)), list(c_intercambios.values()))
plt.xticks(range(len(c_intercambios)), list(c_intercambios.keys()))
plt.xlabel('Numero de intercambios')
plt.ylabel('Cantidad')
plt.title('Intercambios para n = {}'.format(A))    
print('\nGrafica:')
plt.show() 
```
